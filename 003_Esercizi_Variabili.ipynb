Esercizio 6
Incrementare "numero1" di 1 e decrementare "numero2" di 3. Mandare a schermo i nuovi valori.

```python
numero1 = int(input("Inserisci il primo numero: "))
numero2 = int(input("Inserisci il secondo numero: "))

def nuovi_valori (numero1, numero2):
    numero1 += 1
    numero2 -= 3
    return (numero1, numero2)

numero_1_incremento, numero_2_INCREMENTO = nuovi_valori(numero1, numero2)

print(f"Il nuovo numero e': {numero_1_incremento} ")
print(f"Il nuovo numero e' : {numero_2_INCREMENTO}")
```

Esercizio 7
Moltiplicare "numero1" per se stesso + 5, sommarlo quindi a "numero2" elevato alla seconda. Assegnare l'operazione alla variabile risultato e mandare a schermo.

```python
numero_1 = int(input("Inserisci il primo numero: "))
numero_2 = int(input("Inserisci il secondo numero: "))

def calcolo (numero_1, numero_2):
    risultato = (numero_1 * (numero_1 + 5)) + (numero_2 ** 2)
    return risultato

risultato = calcolo(numero_1, numero_2)

print(f"Il risultato e': {risultato}")
```
