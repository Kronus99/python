Esercizio 1
Scrivere un programma che utilizzi un loop for per stampare ogni elemento di una lista.

```python
lista = [1, 2, 3, 4, 5]
for numero in lista:
  print(numero)
```

Esercizio 3
Scrivere un programma che utilizzi un loop for per sommare tutti i numeri in una lista.

```python
lista = [1, 2, 3, 4, 5]
somma = 0
for numero in lista:
  somma += numero
print(somma)
```

```python
somma = 0

for i in range(5):
    numero = int(input(f"Inserisci il numero {i+1}: "))
    somma += numero

print(f"La somma dei numeri inseriti è: {somma}")
```

Scrivere un programma che utilizzi un loop for per stampare tutte le chiavi di un dizionario.

```python
dizionario = {"NOME": 1, "COGNOME": 2, "ETA": 3}
for chiave in dizionario.keys():
  print(chiave)
#
```

Esercizio 7
Scrivere un programma che utilizzi un loop for per stampare tutte le coppie chiave-valore di un dizionario.

```python
dizionario = {"NOME:": "SILVIO", "COGNOME:": "COPPA", "ETA:": 25}
for chiave, valore in dizionario.items():
  print(chiave, valore)
```
